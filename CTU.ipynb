```python
# One-hot encoding of 'Proto' column
onehot_encoder = OneHotEncoder(sparse=False)
proto_onehot = onehot_encoder.fit_transform(df[['Proto']])
proto_df = pd.DataFrame(proto_onehot, columns=onehot_encoder.get_feature_names_out(['Proto']))
df = pd.concat([df, proto_df], axis=1)
df.drop('Proto', axis=1, inplace=True)
```

```python
def increaser(label_col):
    arr = []
    for label in label_col:
        if 'Botnet' in str(label):
            arr.append(1)
        elif 'Background' in str(label):
            arr.append(0.3)
        else:
            arr.append(0)
    return np.array(arr) + df['TotPkts'] * df['TotBytes'] * df['SrcBytes']

df['Label'] = arr = increaser(df['Label'])
```

```python
import ipaddress
import ipaddress

def convert_ip_to_int(ip):
    try:
        # Attempt to convert as IPv4
        return int(ipaddress.IPv4Address(ip))
    except ipaddress.AddressValueError:
        # Convert as IPv6
        return int(ipaddress.IPv6Address(ip))

df['SrcAddr_binary'] = df['SrcAddr'].apply(convert_ip_to_int)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Split the data into X and y
X = df[['Dur', 'Sport', 'Dport', 'State', 'TotPkts', 'TotBytes',
       'SrcBytes', 'Proto_icmp', 'Proto_rtcp', 'Proto_rtp',
       'Proto_tcp', 'Proto_udp', 'Proto_udt', 'Dir_   ->', 'Dir_   ?>',
       'Dir_  <-', 'Dir_  <->', 'Dir_  <?', 'Dir_  <?>', 'SrcAddr_binary']]
y = df['Label']

# Train a Random Forest model to get feature importance scores
rf = RandomForestRegressor(random_state=42)
rf.fit(X, y)

# Get feature importance scores
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

# Print feature importance scores
print("Feature ranking:")
for f in range(X.shape[1]):
    print(f"{f+1}. {X.columns[indices[f]]} ({importances[indices[f]]:.4f})")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
print(X_train.shape,y.shape)
print(X.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split

# Convert the input data to float32
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Create a simple neural network model
model = tf.keras.Sequential([
    Dense(1024, activation='relu', input_shape=(1,)),
    Dropout(0.5),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid'),
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=16, validation_split=0.2)
```

```python
# Predict on test data
y_pred = model.predict(X_test)

# Show scatter plot of y_pred vs. y_true
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel('y_true')
plt.ylabel('y_pred')
plt.title('Scatter plot of y_pred vs. y_true')
plt.show()

# Evaluate the model on the test set
loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Test loss: {loss:.3f}')
```
